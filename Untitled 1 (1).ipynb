```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances,cosine_similarity

# Sample data
data = df
# Create DataFrame
df = pd.DataFrame(data)

# typeConversion
df['COURSENAME'] = df['COURSENAME'].astype(str)
df['SKILLS'] = df['SKILLS'].astype(str)
df['PROJECT_TECHSTACK'] = df['PROJECT_TECHSTACK'].astype(str)


# Concatenate columns into a single string for each row
df['combined'] = df['COURSENAME'] + " " + df['PROJECT_TECHSTACK'] + " " + df['SKILLS']

# Vectorize the combined strings
# Vectorize the combined strings
vectorizer = CountVectorizer()
skill_vectors = vectorizer.fit_transform(df['combined'])

# User input string representing required skills for the project
user_string = "Nodejs"
user_vector = vectorizer.transform([user_string])

# Calculate Euclidean distances between the user vector and each employee vector
euclidean_distances = euclidean_distances(user_vector, skill_vectors)

# Calculate cosine similarity between the user vector and each employee vector
cosine_similarities = cosine_similarity(user_vector, skill_vectors)

# Find the index of the employee with the smallest Euclidean distance
euclidean_recommendation_index = euclidean_distances.argmin()

# Find the index of the employee with the highest cosine similarity
cosine_recommendation_index = cosine_similarities.argmax()

# Get the name of the recommended employee using both Euclidean distance and cosine similarity
euclidean_recommended_employee = df.iloc[euclidean_recommendation_index]['USERID']
cosine_recommended_employee = df.iloc[cosine_recommendation_index]['USERID']

print("Recommended Employee (Euclidean Distance):", euclidean_recommended_employee)
print("Recommended Employee (Cosine Similarity):", cosine_recommended_employee)

```
